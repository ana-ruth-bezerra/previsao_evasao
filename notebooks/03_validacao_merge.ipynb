```python
import pandas as pd
```

```python
df_matriculas = pd.read_csv("../data/processed/microdados_matriculas_2023_reduzido.csv", sep=';')
df_eficiencia = pd.read_csv("../data/processed/microdados_eficiencia_academica_2023_reduzido.csv", sep=';', decimal=',')

chave_curso = ['Instituição', 'Nome de Curso', 'Tipo de Oferta', 'Unidade de Ensino']
df_eficiencia_ref = df_eficiencia.groupby(chave_curso, as_index=False).agg(
    Fator_Esforco_Medio=('Fator Esforço Curso', 'mean'),
    Carga_Horaria=('Carga Horaria', 'first'),
    Modalidade_de_Ensino=('Modalidade de Ensino', 'first')
)

df_merge = pd.merge(df_matriculas, df_eficiencia_ref, on=chave_curso, how='left')
df_secundario = df_merge.dropna(subset=['Fator_Esforco_Medio']).reset_index(drop=True)

df_secundario.to_csv("../data/processed/base_secundaria.csv", sep=';', encoding='utf-8', index=False)

print(f"Base principal : {df_matriculas.shape[0]:>6} registros | {df_matriculas.shape[1]} colunas")
print(f"Base secundária: {df_secundario.shape[0]:>6} registros | {df_secundario.shape[1]} colunas ({len(df_secundario)/len(df_matriculas)*100:.1f}% da principal)")
```
